Step 1: create the dictionary and read the keys

```python
# keys: morse, values: characters
MORSE_TO_CHAR = {}

with open('key.txt', 'r') as f:
  for line in f:
    # split line by space
    words = line.split() # ['A', '.-']
    MORSE_TO_CHAR[ words[1] ] = words[0] # { '.-': 'A'}
  f.close()
```

```python
print(MORSE_TO_CHAR)
```

Step 2: Open the file to decode and decode

```python
# file you want to decode
file = "SEGMENT3-1.txt"
# use f string to put that file in the path
filepath = f"../segments/{file}"

with open(filepath, 'r') as f:
  # convert to a string
  morse_text = f.readline()
  # Normalize word separators: treat ' / ' or '/' as word boundary marker
  morse_text = morse_text.strip().replace(" / ", "   ")
  
	# Split into words (separated by 3 spaces)
  words = morse_text.split("   ")
  
  decoded_words = []
  for word in words:
    decoded_letters = []
    # Split each word into individual morse codes (separated by 1 space)
    codes = word.strip().split(" ")
    for code in codes:
      code = code.strip()
      if code in MORSE_TO_CHAR:
        decoded_letters.append(MORSE_TO_CHAR[code])
    decoded_words.append("".join(decoded_letters))
    
        
print(" ".join(decoded_words))
```
